```python
# Word is actually a word and not a stopword

def remove_stopwords(text):
    stopwords = nltk.corpus.stopwords.words("english")
    content = [word for word in text if word.isalpha() and word.lower() not in stopwords]
    print(content[:50])
    return content
    
remove_stopwords(hobbies_text)
```

```python
fdist2 = nltk.FreqDist(remove_stopwords(hobbies_text))
entries = fdist2.most_common(50)
```
