```python
# Import texts from the Internet

from urllib import request
url = "http://www.gutenberg.org/files/2554/2554-0.txt"
response = request.urlopen(url)
raw = response.read().decode("utf8")
type(raw)
len(raw)
raw[:75]
```

```python
# Tokenization = splitting text into individual words
tokens = word_tokenize(raw)
tokens[:100]
```

```python
# Casting is changing formats. eg: detokenizing
text=nltk.Text(tokens)

# Collocations - words that often occur adjacent to each other
text.collocations()
```

# Regular expressions for pattern matching

```python
wordlist = [word for word in nltk.corpus.words.words("en") if word.islower()]
[word for word in wordlist if re.search("ed$", word)]

# words ending in ed
```

```python
# ^ is the top of the word
# $ is the end of the word
# . is one-letter wildcard

[w for w in wordlist if re.search("^..j..tion$", w)]
```

```python
[w for w in wordlist if re.search("^..tt.ng$", w)]
```

```python
# Like a T9 old phone keypad! 
# Words that can be written by pressing 2278 on a T9 keypad

[w for w in wordlist if re.search("^[abc][abc][tuv][pqrs]$", w)]
```

```python
[w for w in wordlist if re.search("^[g-o]at$", w)]
```

```python
chat_words = sorted(set([w for w in nltk.corpus.nps_chat.words()]))

# + for multiple characters

[w for w in chat_words if re.search("^m+i+n+e+$", w)]
```

```python
# You don't need to specify start or end

[w for w in chat_words if re.search("n+o+", w)]
```

```python
# Require groups of letters with parentheses

[w for w in chat_words if re.search("(uck)", w)]
```

```python
# ^ inside brackets means not
[w for w in wordlist if re.search("^.[^e]tt.ng$", w)]
```
